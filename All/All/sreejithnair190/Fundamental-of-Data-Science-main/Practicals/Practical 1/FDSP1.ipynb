```python
from pandas import DataFrame
Cars={
    'Brand':['BMW','Honda','Ford','Audi'],
    'Price':[23121,43534,21432,21421]
}
df=DataFrame(Cars,columns=['Brand','Price'])
df
```

```python
con=sqlite3.connect('CARS_DB.db')
c=con.cursor()
```

```python
c.execute('CREATE TABLE CARS(BRAND text,PRICE number)')
con.commit()
```

```python
df.to_sql('CARS',con,if_exists='replace',index=False)
```

```python
c.execute('SELECT Brand,max(Price) from CARS')
```

```python
df=DataFrame(c.fetchall(),columns=['Brand','Price'])
df
```
